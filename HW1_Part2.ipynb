```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.DataFrame(pd.read_csv('StudentsPerformance.csv'))
```

```python
mean=data.groupby(data["race/ethnicity"], as_index=False).agg({"math score" : "mean","reading score" : "mean","writing score" : "mean"})
```

```python
mean
```

```python
plt.plot(mean['race/ethnicity'],mean['math score'])
plt.plot(mean['race/ethnicity'],mean['reading score'])
plt.plot(mean['race/ethnicity'],mean['writing score'])
```

```python
sns.heatmap(data.corr(), vmax=0.6, square=True, annot=True)
```

```python
mean_gender=data.groupby(data["gender"], as_index=False).agg({"math score" : "mean","reading score" : "mean","writing score" : "mean"})
```

```python
gender = ["female", "male"]
mean_gender.set_index("gender").loc[gender].plot(kind="bar", legend=False)
```

```python
mean_edu=data.groupby(data["parental level of education"], as_index=False).agg({"math score" : "mean","reading score" : "mean","writing score" : "mean"})
```

```python
plt.figure(figsize=(12,10))
plt.plot(mean_edu['parental level of education'],mean_edu['math score'])
plt.plot(mean_edu['parental level of education'],mean_edu['reading score'])
plt.plot(mean_edu['parental level of education'],mean_edu['writing score'])
```
